```python
# Dependencies
import pandas as pd
import numpy as np
from os import path
from PIL import Image
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import nltk 
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Load and read CSV file
file = 'list_gz.csv'
df = pd.read_csv(file, index_col=0, encoding="ISO-8859-1")
df.head()
```

```python
pd.set_option('display.max_rows', 110)
df.dtypes
```

```python
df = df.rename(columns={'neighbourhood_cleansed': 'Chicago Neighbourhoods', 'listing_url': 'AirBnB Listings', 'review_scores_rating': 'Avg Reviews Ratings','reviews_per_month': 'Reviews Per Month'})
```

# ChiTown AirBnB Rentals

```python
TopNeighbourhoods = pd.DataFrame(df.groupby('Chicago Neighbourhoods')['AirBnB Listings'].nunique())
TopNeighbourhoods.sort_values(by=['AirBnB Listings'], ascending=False, inplace=True)
TopNeighbourhoods.head(5)
```

Chicago Neighbourhoods with over 500 AirBnB listings

```python
BottomNeighbourhoods = pd.DataFrame(df.groupby('Chicago Neighbourhoods')['AirBnB Listings'].nunique())
BottomNeighbourhoods.sort_values(by=['AirBnB Listings'], ascending=True, inplace=True)
BottomNeighbourhoods.head(10)
```

*Chicago Neighbourhoods with tmhe less than 5 AirBnB listings

```python
Monthly_Reviews = pd.DataFrame(df.groupby('Chicago Neighbourhoods')['Reviews Per Month'].nunique())
Monthly_Reviews.sort_values(by=['Reviews Per Month'], ascending=False, inplace=True)
Monthly_Reviews.head(5)
```

*Neighbourhoods with the most reviews!
